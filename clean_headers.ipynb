```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import the life expectancy dataset
df = pd.read_csv('Life_Expectancy_Data.csv')

# before cleaning
df.head()
```

```python
# clean headers of the dataset, remove spaces found at the beginning and end of the column names
df.columns = df.columns.str.strip()
df.head()
```

```python
# Capitalize the first letter of each word in the column names
df.columns = df.columns.str.title()
df.head()
```

```python
# Fix spelling of Diphtheria
df.rename(columns={'Diphtheria': 'Diptheria'}, inplace=True)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = df

# Plot the relationship between each individual immunization and life expectancy
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.scatter(data["Diptheria"], data["Life Expectancy"], s=1, c="red", alpha=0.5)
plt.xlabel("Diphtheria Immunization Coverage (%)")
plt.ylabel("Life Expectancy (Years)")
# show trend line ?

plt.subplot(1, 3, 2)
plt.scatter(data["Polio"], data["Life Expectancy"], s=1, c="green", alpha=0.5)
plt.xlabel("Polio Immunization Coverage (%)")
plt.ylabel("Life Expectancy (Years)")
plt.title("Relationship between Immunization and Life Expectancy")

plt.subplot(1, 3, 3)
plt.scatter(data["Hepatitis B"], data["Life Expectancy"], s=1, c="blue", alpha=0.5)
plt.xlabel("Hepatitis B Immunization Coverage (%)")
plt.ylabel("Life Expectancy (Years)")

plt.tight_layout()
plt.show()
```
